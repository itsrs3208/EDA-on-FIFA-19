## What is EDA ? <br>
#### Exploratory Data Analysis (EDA) is the process of examining a dataset to uncover patterns, relationships, and insights. It involves understanding the structure of the data, handling missing values, and visualizing distributions using techniques like histograms, scatter plots, and correlation heatmaps. EDA helps identify outliers, trends, and anomalies, providing a foundation for hypothesis testing. Feature engineering is often a key part of EDA, where new variables are created to improve model performance. Ultimately, EDA ensures data quality and provides insights for building predictive or explanatory models.

## Contents of the dataset:
ID    - Unique Identity number given to each player<br>
Overall - Approx rating of a player based on their stats and performance <br>
Potential - Approx rating a player could achieve <br>
Club - A player's league team name<br>
Value - Market value of the player<br>
Position - Preffered position that the player plays <br>
Release Clause - A specific amount of money to be paid to the player owning club for termination of contract before the specified period of time

# DATA VIZUALIZATION 

#### Q:What does this bar plot of player position tells us ? <br>
~ we can clearly see that striker(ST) position has the most players while Left forward(LF) has the least amount of players.

```python
df['Position'].value_counts().plot(kind='barh')
plt.title('Player count in each position')
```

#### Q:Which foot is better preffered by majority population of the player? <br>
 ~From the below countplot we can clearly see that most of the players prefer right foot over left foot 

```python
sns.countplot(x='Preferred Foot',data=df)
plt.title('Most preffered foot ')
```

#### Q:Which country has the most amount of players in the dataset?<br>
~England has the most amount of players (28.74%) followed by Germany,Spain,Argentina and France

```python
country_names=df.Nationality .value_counts().index
```

```python
country_values=df.Nationality .value_counts().values
```

```python
explode=[0.1,0,0,0,0]
plt.pie(country_values[:5],labels=country_names[:5],explode=explode,shadow =True,autopct='%.2f%%')
plt.show()
```

#### Height and weight range of players in the dataset

```python
sns.boxenplot(df['Height'])
plt.ylabel('Height in feet')
plt.title('Trends among player heights')
```

```python
sns.boxenplot(df['Weight'])
plt.ylabel('Weight in pound')
plt.title('Trends among player weights')
```

#### Distribution of Age 

```python
sns.set(style="whitegrid")
sns.displot(x='Age',kind='kde',data=df)
plt.title('Age distribution of players')
```

#### Q: Who are the top 10 most valued players in the dataset?

#### Q: Who are the top 10 players with the highest release clause in the dataset?

#### Q: Who are the top 10 players with the highest wages in the dataset?

```python
df_wage = df.sort_values(by='Wage',ascending=False).head(10)
```

#### Heatmap data of the dataset

```python
corr=df.drop(columns=['ID', 'Name','Nationality','Club','Joined','Contract Valid Until','Preferred Foot','Position']).corr()
plt.figure(figsize=(14,9))
sns.heatmap(corr,annot=True)
```

## Inference from the corelation matrix :

• Age and Overall : A moderate but positive correlation(0.45) exists between age and overall.Thus older player tend to have higher overall but it's     not a strong conclusion<br>
• Age and Potential : A negative correlation(-0.25) exists between them , thus younger players tend to have higher potential<br>
• Overall and Wage/Value : A stronger positive correlation exists between Overall and Wage , Value. Thus players with greater overall will have a       better market value as well as better wages<br>
• Overall and Potential : A strong positive correlation(0.66) exists between them which denotes that players with higher overall tends to have greater   potentials<br>
• Potential and Wage/Value : A stronger positive correlation exists between potential and Wage,Value.<br>
• Value and Release clause : A really strong positive correlation(0.97) exists between Value and Release clause.Thus teams dont want their highly       valued players to leave their team before their contract ends.<br>
• Skill Moves and Height/Weight: Skill moves have a negative correlation with height (-0.42) and weight (-0.35). Players with more skill moves tend to   be shorter and lighter.<br>
• International Reputation and Age: A moderate positive correlation (0.25) suggests that players with higher international reputations are slightly     older.
